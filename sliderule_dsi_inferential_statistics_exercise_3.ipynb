# Hospital Readmissions Data Analysis and Recommendations for Reduction

### Background
In October 2012, the US government's Center for Medicare and Medicaid Services (CMS) began reducing Medicare payments for Inpatient Prospective Payment System hospitals with excess readmissions. Excess readmissions are measured by a ratio, by dividing a hospital’s number of “predicted” 30-day readmissions for heart attack, heart failure, and pneumonia by the number that would be “expected,” based on an average hospital with similar patients. A ratio greater than 1 indicates excess readmissions.

# A - Critique

I do not agree with the analysis due to the following reasons:

* The probabilistic nature of the data was not accounted for. If this study was repeated the results could have been different. The study should have provided margin of error in their conclusions.


* The number of datapoints for larger hospitals are significantly lower than those available for smaller hospitals. As a result, the mean excess readmission rate may appear lower for them. If there were equal number of datapoints for both sizes of hospitals, then may be one could have seen higher mean excess readmission rate for the larger hopsitals as well.  


* For measuring the correlation between the hospital size and readmission rate, pearson's coefficient (or) the slope of regression analysis should have been used.  


* In the analysis below, linear regression model was applied to the data. The slope of this line and its confidence interval were calculated. Hypothesis testing shows that there is slight slope, approx -2.8x10^-5 (between -2x10^-5 and -4x10^-5 ). In other words, when the hospital size/number of discharges increase by four fold (i.e. 10000 times), the mean excess readmission rate lowers by 0.28 points. This number clearly is not zero and was validated by Hypothesis testing. **However, it is also very small effect**. The same effect mentioned above is also seen in the correlation coefficient.  


* **The impact of hospital size on the mean excess readmission rates is very small. Requiring smaller hospitals to upgrade based on such small impact (as evidenced by regression slope) is not the correct decision.**

## B.1 Defining Functions needed

```python
def make_histogram(data, x_label, y_label, fig_no, kde_var=False, **kwargs):
    plt.figure(fig_no, figsize=(8,5))
    #no_of_bins = int(len(data)**0.5) # Calculate bins sqrt(len(data))
    no_of_bins = 30
    x = sns.distplot(data, hist=True, kde=kde_var, rug=False)
    _ = plt.xlabel(x_label)
    _ = plt.ylabel(y_label)
    if 'xlimits' in kwargs:
        _ = plt.xlim(kwargs['xlimits'])
    else:
        plt.margins(0.02)
    
    if 'title' in kwargs:
        _ = plt.title(kwargs['title'], fontsize=16)
    
    plt.axvline(np.mean(data), color='k', linestyle='dashed', linewidth=1)
    plt.axvline(np.median(data), color='r', linestyle='dashed', linewidth=1)
    
     
    ymin, ymax = plt.ylim()
    xmin, xmax = plt.xlim()
    yrange = ymax - ymin
    xrange = xmax - xmin
    mean_string = "Mean (black line) is {}".format(round(np.mean(data),2))
    median_string = "Median (red line) is {}".format(round(np.median(data),2))
    std_string = "Std. Dev is {}".format(round(np.std(data),2))
    plt.annotate(mean_string, xy = (np.mean(data), ymax), xytext = (xmin+xrange*0.7, ymin+yrange*0.7))
    plt.annotate(median_string, xy = (np.median(data), ymax), xytext = (xmin+xrange*0.7, ymin+yrange*0.6))
    plt.annotate(std_string, xy = (np.median(data), ymax), xytext = (xmin+xrange*0.7, ymin+yrange*0.5))
    
    if 'obs_statistic' in kwargs:
        plt.axvline(kwargs['obs_statistic'], color='b', linestyle='dashed', linewidth=1)
        ymin, ymax = plt.ylim()
        xmin, xmax = plt.xlim()
        yrange = ymax - ymin
        xrange = xmax - xmin
        #plt.axvline(0-kwargs['obs_statistic'], color='b', linestyle='dashed', linewidth=1)
        plt.annotate('Observed Statistic in Blue Line', xy = (xmin+xrange*0.1, ymin+yrange*0.5), 
                     xytext = (kwargs['obs_statistic']+xrange*0.1, ymin+yrange*0.25))
   
    plt.show()
```

## B.2 Drawing Boot Strap Samples of regression parameters & Corr Coeff for Hypothesis Testing

```python
def draw_bs_linreg(x, y, size=1):
    """Perform bootstrap for linear regression."""
    
    if (isinstance(x,list)):
        x = np.array(x) # Subsetting using vector of indiex values \
                        # doesn't work on lists. This code converts any lists to numpy array's.
    if (isinstance(y,list)):
        y = np.array(y)
    
    # Initialize replicates: bs_slope_reps, bs_intercept_reps
    bs_slope_reps = np.empty(size)
    bs_intercept_reps = np.empty(size)
    corr_coeff_bs = np.empty(size)

    # Generate replicates
    for i in range(size):
        bs_x = np.random.choice(x, size=len(x))
        bs_y = np.random.choice(y, size=len(y))
        bs_slope_reps[i], bs_intercept_reps[i] =np.polyfit(bs_x, bs_y, 1)
        corr_coeff_bs[i]=np.corrcoef(bs_x, bs_y)[0,1]

    return bs_slope_reps, bs_intercept_reps, corr_coeff_bs

def hypothesis_test(stat_bs, stat_obs, alpha, no_of_runs, null_hypo_string=''):
    p_value = np.sum(stat_bs<=stat_obs)/no_of_runs
    if (p_value < alpha):
        print('p-value= {0}% is <= alpha {1}%, Reject Null Hypothesis: {2}'.format(round(p_value,5)*100, alpha*100,\
                                                                                   null_hypo_string))
    else:
        print('p-value= {0}% is > alpha {1}%, Accept Null Hypothesis: {2}'.format(round(p_value,5)*100, alpha*100,\
                                                                                   null_hypo_string))
```

### Defining parameters & Draw Bootstrap samples of slope & covariance

```python
alpha = 0.01
no_of_runs = 1000
confidence_interval = [0.5, 99.5]
slope_bs, intercept_bs, corr_coeff_bs = draw_bs_linreg(x, y, size=no_of_runs)
```

## B.3 Hypothesis Testing Using Boot Strap Samples of correlation coeff

Statistic: Pearsons coefficient between hospital size (as measured by number of discharges) and mean excess readmission rate.

H0: Pearsons coeff is 0.  
H1: Pearsons coeff is not 0.

Create bootstrap samples of the dataset to test the probability of finding observed pearsons coeff. This would be a one tailed test.

```python
corr_coeff_obs = np.corrcoef(x, y)[0,1]
null_hypo_string = 'There is no correlation between hospital size and mean excess readmission rate'
hypothesis_test(corr_coeff_bs, corr_coeff_obs, alpha, no_of_runs, null_hypo_string)
make_histogram(corr_coeff_bs, 'Perasons Corr Coeff', 'Count', 2, kde_var=True, obs_statistic=corr_coeff_obs)
```

## B.4 Hypothesis Test Using Slope of the Regression Line

Statistic: Slope of the regression line between hospital size (as measured by number of discharges) and mean excess readmission rate.  

H0: Slope of the regression line between hospital size and mean excess readmission rate is zero.  
H1: Slope of the regression line between hospital size and mean excess readmission rate is not zero.

Use bootstrapping of dataset to test the probability of finding observed pearsons coeff. This would be a one tailed test.

```python
slope_obs, intercept_obs = np.polyfit(x, y, 1)
null_hypo_string = 'Slope of the regression line between hospital size and mean excess readmission rate is zero'
hypothesis_test(slope_bs, slope_obs, alpha, no_of_runs, null_hypo_string)
title='Bootstrap samples of Slope of Regression Line Between Hospital Discharges & Mean Excess Readmission Rate'
make_histogram(slope_bs, 'Slope', 'Count', 1, kde_var=True, obs_statistic=slope_obs, title=title)
```

## B.5 Calculating Confidence Interval of the slope & Corr. Coeff

```python
def draw_bs_pairs_linreg(x, y, size=1):
    """Perform pairs bootstrap for linear regression."""
    
    if (isinstance(x,list)):
        x = np.array(x) # Subsetting using vector of indiex values \
                        # doesn't work on lists. This code converts any lists to numpy array's.
    if (isinstance(y,list)):
        y = np.array(y)
    
    # Set up array of indices to sample from: inds
    inds = np.arange(len(x))

    # Initialize replicates: bs_slope_reps, bs_intercept_reps
    bs_slope_reps = np.empty(size)
    bs_intercept_reps = np.empty(size)
    corr_coeff_bs = np.empty(size)

    # Generate replicates
    for i in range(size):
        bs_inds = np.random.choice(inds, size=len(inds))
        bs_x, bs_y = x[bs_inds], y[bs_inds]
        bs_slope_reps[i], bs_intercept_reps[i] =np.polyfit(bs_x, bs_y, 1)
        corr_coeff_bs[i] = np.corrcoef([bs_x,bs_y])[0,1]

    return bs_slope_reps, bs_intercept_reps, corr_coeff_bs


slope_bs_conf, intercept_bs_conf, corr_coeff_conf = draw_bs_pairs_linreg(x, y, size=no_of_runs)
slope_conf_interval = np.percentile(slope_bs_conf, confidence_interval)
corr_coeff_interval = np.percentile(corr_coeff_conf, confidence_interval)
print('slope is in the interval [{0}, {1}]'.format(round(slope_conf_interval[0],5), round(slope_conf_interval[1],5)))
print('Corr Coeff is in the interval [{0}, {1}]'.format(round(corr_coeff_interval[0],5), round(corr_coeff_interval[1],5)))

title='Slope of Regression Line Between Hospital Discharges & Mean Excess Readmission Rate'
make_histogram(slope_bs_conf, 'Slope', 'Count', 1, kde_var=True, obs_statistic=slope_obs, title=title)
title='Corr Coeff Between Hospital Discharges & Mean Excess Readmission Rate'
make_histogram(corr_coeff_conf, 'Corr Coeff', 'Count', 1, kde_var=True, obs_statistic=corr_coeff_obs, title=title)
```
